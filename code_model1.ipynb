```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from PIL import Image
import io

# Load Parquet files
def load_parquet_data(file_path):
    df = pd.read_parquet(file_path)
    return df

def extract_data(df):
    images = []
    valid_labels = []
    for idx, img in enumerate(df['image']):
        try:
            if isinstance(img, dict):  # If the image is stored as a dictionary
                print(f"Debug: Entry at index {idx} is a dictionary.")
                img_data = img.get('bytes', None)
                if img_data is not None:
                    # Decode the image bytes
                    img = Image.open(io.BytesIO(img_data)).convert('RGB')
                    img = np.array(img)  # Convert PIL image to NumPy array
                else:
                    print(f"Warning: Missing 'bytes' key in image dictionary at index {idx}")
                    continue
            elif isinstance(img, np.ndarray):
                print(f"Debug: Entry at index {idx} is an ndarray.")
                img = img  # Use it directly if already a NumPy array
            else:
                print(f"Warning: Unsupported image type: {type(img)} at index {idx}")
                continue

            if len(img.shape) == 3:  # Ensure valid shapes (H, W, C)
                images.append(img)
                valid_labels.append(df['label'].iloc[idx])
            else:
                print(f"Warning: Invalid image shape {img.shape} at index {idx}")
        except Exception as e:
            print(f"Error processing image at index {idx}: {e}")

    return images, valid_labels

# File paths
train_file = "/Users/mariyajose/Documents/weekend_projects/alzheimers_classification/Alzheimer MRI Disease Classification Dataset/Data/train-00000-of-00001-c08a401c53fe5312.parquet"
test_file = "/Users/mariyajose/Documents/weekend_projects/alzheimers_classification/Alzheimer MRI Disease Classification Dataset/Data/test-00000-of-00001-44110b9df98c5585.parquet"

# Load and extract data
train_df = load_parquet_data(train_file)
test_df = load_parquet_data(test_file)

print("Inspecting training data...")
train_images, train_labels = extract_data(train_df)

print("Inspecting test data...")
test_images, test_labels = extract_data(test_df)

# Check for valid data
if len(train_images) == 0:
    raise ValueError("No valid training images found. Please check the dataset.")
if len(test_images) == 0:
    raise ValueError("No valid test images found. Please check the dataset.")

# Verify data
print(f"Number of training images: {len(train_images)}")
if len(train_images) > 0:
    print(f"Shape of first training image: {train_images[0].shape}")

# Normalize and resize images
train_images = np.array([tf.image.resize(img, (224, 224)).numpy() for img in train_images]).astype('float32') / 255.0
test_images = np.array([tf.image.resize(img, (224, 224)).numpy() for img in test_images]).astype('float32') / 255.0

# One-hot encode labels
num_classes = len(np.unique(train_labels))
train_labels = tf.keras.utils.to_categorical(train_labels, num_classes)
test_labels = tf.keras.utils.to_categorical(test_labels, num_classes)

# Split train set into train and validation
X_train, X_val, y_train, y_val = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Data augmentation
data_gen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Model definition using transfer learning
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
out = Dense(num_classes, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=out)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(
    data_gen.flow(X_train, y_train, batch_size=32),
    validation_data=(X_val, y_val),
    epochs=10
)

# Evaluate on the test set
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Classification report
predictions = model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(test_labels, axis=1)

print("Classification Report:")
print(classification_report(true_classes, predicted_classes, target_names=['Mild_Demented', 'Moderate_Demented', 'Non_Demented', 'Very_Mild_Demented']))

# Confusion Matrix
conf_matrix = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(8, 6))
plt.title("Confusion Matrix")
plt.imshow(conf_matrix, interpolation='nearest', cmap='Blues')
plt.colorbar()
plt.xticks(range(num_classes), ['Mild', 'Moderate', 'Non', 'Very Mild'], rotation=45)
plt.yticks(range(num_classes), ['Mild', 'Moderate', 'Non', 'Very Mild'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Save the model
model.save("alzheimers_mri_model.h5")

print("Model training and evaluation completed.")
```
